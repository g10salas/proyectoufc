# 📓 01. Exploración Inicial de Datos Históricos de la UFC

Este notebook cumple con la fase de **Exploración de Datos** dentro de nuestro ecosistema. Su objetivo es conectarse a la base de datos analítica **DuckDB**, verificar las tablas disponibles y hacer una inspección rápida de las primeras filas de cada entidad (`events`, `fighters`, `fighters_stats` y `fights`) para entender el esquema de datos.

```python
import pandas as pd
from mma_project.data.loader import get_engine

# Cargamos el motor analítico centralizado
engine = get_engine()
print("✅ Conectado a DuckDB para la exploración inicial.")
```

```python
# Consulta estándar para listar todas las tablas creadas en el esquema principal
query_tables = "SELECT table_name FROM information_schema.tables WHERE table_schema = 'main';"

with engine.connect() as conn:
    tablas = pd.read_sql(query_tables, conn)

print("📋 Tablas encontradas en la base de datos analítica:")
print(tablas)
```

```python
print("👤 Muestra de la tabla de Peleadores (fighters):")
with engine.connect() as conn:
    df_fighters = pd.read_sql("SELECT * FROM fighters LIMIT 3;", conn)
display(df_fighters)

print("\n📅 Muestra de la tabla de Eventos (events):")
with engine.connect() as conn:
    df_events = pd.read_sql("SELECT * FROM events LIMIT 3;", conn)
display(df_events)
```

```python
print("🥊 Muestra de la tabla de Peleas (fights):")
with engine.connect() as conn:
    df_fights = pd.read_sql("SELECT * FROM fights LIMIT 3;", conn)
display(df_fights)

print("\n📈 Muestra de la tabla de Estadísticas Acumuladas (fighters_stats):")
with engine.connect() as conn:
    df_stats = pd.read_sql("SELECT * FROM fighters_stats LIMIT 3;", conn)
display(df_stats)
```
